```python
class Pessoa():
    def __init__(self):
        self.nome = ''
        self.idade = 0
    
    def desc(self):
        print(f'{self.nome} tem {self.idade} anos.')
    

pessoa_1 = Pessoa()
pessoa_1.nome = 'Jorge'
pessoa_1.idade = 30

pessoa_2 = Pessoa()
pessoa_2.nome = 'Valéria'
pessoa_2.idade = 27

pessoa_1.desc()
pessoa_2.desc()
```

```python
class Conta():
    def __init__(self):
        self.titular = ''
        self.saldo = 0.0
    
    def depositar(self, valor):
        self.saldo += valor
        
    def sacar(self, valor):
        if self.saldo >= valor:
            self.saldo -= valor
        else:
            print('Saldo insuficiente')
            
    def ver_saldo(self):
        print(f'Seu saldo é de R${self.saldo}')
    
            
conta = Conta()
conta.titular = 'Jorge'
conta.saldo = 100

conta.depositar(100)
conta.sacar(200)
conta.ver_saldo()
```

```python
class Retangulo():
    def __init__(self):
        self.altura = 0.0
        self.largura = 0.0
        
    def ver_perimetro(self):
        print(f'O perímetro do retângulo é {2 * (self.largura + self.altura)}m2')
        
    def ver_area(self):
        print(f'O perímetro do retângulo é {self.largura * self.altura}m')
        

retangulo = Retangulo()
retangulo.altura = 10
retangulo.largura = 10
retangulo.ver_area()
retangulo.ver_perimetro()
```

```python
class Aluno():
    def __init__(self):
        self.nome = ''
        self.notas = []
    
    def calcular_media(self):
        print(f'A média do aluno é de {sum(self.notas) / len(self.notas)}')
        

aluno = Aluno()
aluno.nome = 'Jorge'
aluno.notas = [10, 10, 10]
aluno.calcular_media()
```

```python
class Produto():
    def __init__(self):
        self.nome = ''
        self.preco = 0.0
        
    def info(self):
        print(f'Produto: {self.nome}, Preço: R${self.preco:.2f}')
    
    
class Estoque():
    def __init__(self):
        self.estoque = []
    
    def adicionar_produto(self, produto):
        if isinstance(produto, Produto):
            self.estoque.append(produto.nome)
            print(f'Produto "{produto.nome}" adicionado ao estoque.')
        else:
            print('Apenas objetos da classe "Produto" podem ser adicionados.')
        
    def ver_produtos(self):
            if not self.estoque:
                print('O estoque está vazio.')
            else:
                print('Produtos em estoque:')
                for produto in self.estoque:
                    print(produto)
        

produto_1 = Produto()
produto_1.nome = 'A'
produto_1.preco = 10

produto_2 = Produto()
produto_2.nome = 'B'
produto_2.preco = 20
        
estoque = Estoque()
estoque.adicionar_produto(produto_1)
estoque.adicionar_produto(produto_2)
estoque.ver_produtos()
```

```python
class Livro():
    def __init__(self):
        self.titulo = ''
        self.autor = ''
    
    def info(self):
        print(f'{self.titulo}, {self.autor}')

class Biblioteca():
    def __init__(self):
        self.acervo = []
        
    def adicionar_livro(self, livro):
        if isinstance(livro, Livro):
            self.acervo.append(livro.titulo)
            print(f'Livro "{livro.titulo}" adicionado ao acervo.')
        else:
            print('Apenas objetos da classe "Livro" podem ser adicionados.')
            
    def ver_acervo(self):
        print('Acervo:')
        for livro in self.acervo:
            print(livro)
    
        
livro_1 = Livro()
livro_1.titulo = 'Crime e Castigo'
livro_1.autor = 'Fiódor Dostoiévski'

livro_2 = Livro()
livro_2.titulo = 'Os Irmãos Karamazov'
livro_2.autor = 'Fiódor Dostoiévski'
        
biblioteca = Biblioteca()
biblioteca.adicionar_livro(livro_1)
biblioteca.adicionar_livro(livro_2)
biblioteca.ver_acervo()
```

```python
class Contato():
    def __init__(self):
        self.nome = ''
        self.telefone = ''
        
    def info(self):
         print(f'Contato: {self.nome} | Telefone: {self.telefone}')
         

class Agenda():
    def __init__(self):
        self.contatos = {}
        
    def adicionar_contato(self, contato):
        if isinstance(contato, Contato):
            self.contatos[contato.nome] =  contato.telefone
            print(f'{contato.nome} adicionado à agenda.')
        else:
            print('Apenas objetos da classe "Contato" podem ser adicionados.')
    
    def buscar_contato(self, nome):
        if self.contatos.get(nome):
            print(f'Telefone de {nome}: {self.contatos[nome]}')
        else:
            print('Contato inexistente')
    
    def ver_contatos(self):
        print('Contatos:')
        for nome, telefone in self.contatos.items():
            print(f'Nome: {nome} | Telefone: {telefone}')
            

contato_1 = Contato()
contato_1.nome = 'Jorge'
contato_1.telefone = '(11) 91234-1234'

contato_2 = Contato()
contato_2.nome = 'Valeria'
contato_2.telefone = '(11) 91234-1234'
        
agenda = Agenda()
agenda.adicionar_contato(contato_1)
agenda.adicionar_contato(contato_2)

agenda.buscar_contato('Jorge')
agenda.buscar_contato('Danielle')
agenda.ver_contatos()
```

```python
class Item():
    def __init__(self):
        self.nome = ''
        self.preco = 0.0
    
    def info(self):
         print(f'Produto: {self.nome} | Preço: {self.preco}')
         

class Carrinho():
    def __init__(self):
        self.items = {}
    
    def adicionar_item(self, item):
        if isinstance(item, Item):
            self.items[item.nome] = item.preco
        else:
            print('Apenas objetos da classe "Item" podem ser adicionados.')
    
    def calcular_compra(self):
        print(f'Valor da compra: R${sum(self.items.values())}')
        
    
produto_1 = Item()
produto_1.nome = 'A'
produto_1.preco = 10

produto_2 = Item()
produto_2.nome = 'B'
produto_2.preco = 20
        
carrinho = Carrinho()
carrinho.adicionar_item(produto_1)
carrinho.adicionar_item(produto_2)
carrinho.adicionar_item('item')
carrinho.calcular_compra()
```

```python
import random

class Forca:
    def __init__(self, palavras, tentativas=6):
        self.palavra_secreta = random.choice(palavras).upper()
        self.letras_corretas = set()
        self.letras_erradas = set()
        self.tentativas_restantes = tentativas
        self.total_tentativas = tentativas

    def _mostrar_forca(self):
        estagios = [
            '''
              +---+
              |   |
                  |
                  |
                  |
                  |
            =========
            ''',
            '''
              +---+
              |   |
              O   |
                  |
                  |
                  |
            =========
            ''',
            '''
              +---+
              |   |
              O   |
              |   |
                  |
                  |
            =========
            ''',
            '''
              +---+
              |   |
              O   |
             /|   |
                  |
                  |
            =========
            ''',
            '''
              +---+
              |   |
              O   |
             /|\\  |
                  |
                  |
            =========
            ''',
            '''
              +---+
              |   |
              O   |
             /|\\  |
             /    |
                  |
            =========
            ''',
            '''
              +---+
              |   |
              O   |
             /|\\  |
             / \\  |
                  |
            =========
            '''
        ]
        return estagios[self.total_tentativas - self.tentativas_restantes]

    def _mostrar_palavra(self):
        resultado = ''
        for letra in self.palavra_secreta:
            if letra in self.letras_corretas:
                resultado += letra + ' '
            else:
                resultado += '_ '
        return resultado.strip()

    def verificar_letra(self, letra):
        letra = letra.upper()
        if len(letra) != 1 or not letra.isalpha():
            print('Entrada inválida. Digite apenas uma letra.')
            return

        if letra in self.letras_corretas or letra in self.letras_erradas:
            print(f'Você já tentou a letra "{letra}".')
            return

        if letra in self.palavra_secreta:
            self.letras_corretas.add(letra)
            print(f'Parabéns! A letra "{letra}" está na palavra.')
        else:
            self.letras_erradas.add(letra)
            self.tentativas_restantes -= 1
            print(f'Que pena! A letra "{letra}" não está na palavra.')
            
    def jogar(self):
        print('Bem-vindo ao Jogo da Forca!')
        print('Adivinhe a palavra secreta, uma letra por vez.')
        
        while self.tentativas_restantes > 0:
            print(self._mostrar_forca())
            print(f'\nPalavra: {self._mostrar_palavra()}')
            print(f'Tentativas restantes: {self.tentativas_restantes}')
            print(f'Letras erradas: {", ".join(sorted(list(self.letras_erradas)))}')
            
            letra = input('Digite uma letra: ').strip()
            self.verificar_letra(letra)
            
            if set(self.palavra_secreta).issubset(self.letras_corretas):
                print(f'\nParabéns! Você adivinhou a palavra: {self.palavra_secreta}')
                return

        print(self._mostrar_forca())
        print(f'\nSuas tentativas acabaram! A palavra secreta era: {self.palavra_secreta}')

lista_de_palavras = ['PYTHON', 'PROGRAMACAO', 'DESENVOLVIMENTO', 'ALGORITMO']
jogo = Forca(lista_de_palavras, tentativas=6)
jogo.jogar()
```
